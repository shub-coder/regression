```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
y=df['TAX']
x=df['MEDV']
sns.scatterplot(x,y)
plt.show()
```

```python
a=x.mean()
a
```

```python
b=y.mean()
b
```

```python
m=len(x)
m
```

```python
numer=0
denom=0
for i in range(m):
    numer+=(x[i]-a)*(y[i]-b)
    denom+=(x[i]-a)**2
    print(numer)
b1=numer/denom
b0=b-(b1*a)
print(b1,b0)
```

```python
max_X=np.max(x)+100
max_X
```

```python
min_X=np.min(x)-100
min_X
```

```python
X=np.linspace(min_X,max_X,1000)
Y=b0+b1*X
print(X)
Y
```

```python
plt.plot(X,Y,color='red',label='regression line')
plt.scatter(x,y,color='orange',label='scatter plot')
plt.xlabel('MEDV')
plt.ylabel('TAX')
plt.legend()
plt.show()
```

```python
ss_t=0
ss_r=0
for i in range(m):
    y_pred=b0+(b1*x[i])
    ss_t+=(y[i]-b)**2
    ss_r+=(y[i]-y_pred)**2
r2=1-(ss_r/ss_t)
r2
```
